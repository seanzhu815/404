```python
import statsmodels.api as sm
import statsmodels.formula.api as smf
```

```python

df.rename(columns={"Lo 10": "lo10"},inplace=True)
df.rename(columns={"Hi 10": "hi10"},inplace=True)
df.rename(columns={"decile-2": "decile2"},inplace=True)
df.rename(columns={"decile-3": "decile3"},inplace=True)
df.rename(columns={"decile-4": "decile4"},inplace=True)
df.rename(columns={"decile-5": "decile5"},inplace=True)
df.rename(columns={"decile-6": "decile6"},inplace=True)
df.rename(columns={"decile-7": "decile7"},inplace=True)
df.rename(columns={"decile-8": "decile8"},inplace=True)
df.rename(columns={"decile-9": "decile9"},inplace=True)
df['lo10_lag']=df['lo10'].shift(1)
df['decile3_lag']=df['decile3'].shift(1)
df['decile4_lag']=df['decile4'].shift(1)
df['decile5_lag']=df['decile5'].shift(1)
df['decile6_lag']=df['decile6'].shift(1)
df['decile7_lag']=df['decile7'].shift(1)
df['decile8_lag']=df['decile8'].shift(1)
df['decile9_lag']=df['decile9'].shift(1)
df['decile2_lag']=df['decile2'].shift(1)
df['hi10_lag']=df['hi10'].shift(1)
df['ewretd_lag']=df['ewretd'].shift(1)
```
